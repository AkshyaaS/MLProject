APPLYING LOGISTIC REGRESSION


```python
#Setting input and target features
X = df.drop(columns = ['date','Occupancy'])
Y = df['Occupancy']
```

```python
from sklearn.preprocessing import StandardScaler
scaled_data = StandardScaler().fit_transform(X)
```

```python
X_scaled = pd.DataFrame(scaled_data,columns = X.columns)  #Standardizing the input data
```

```python
X_scaled  #Standardized input data
```

```python
X = X_scaled
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
LR = LogisticRegression()
```

```python
#Analyzing the accuracy values for various cross validation values
for i in range(2,10,1):
  CV_Score = cross_val_score(LR,X,Y,scoring = 'accuracy',cv = i).mean()
  print('cv = ',i,'Accuracy = ',CV_Score)
```

Maximum Accuracy occurs when the cross validation = 5

```python
from sklearn.model_selection import train_test_split
X_test,X_rem,y_test,y_rem = train_test_split(X, Y,test_size=0.1,random_state=100)
```

```python
#Analyzing the accuracy values for various cross validation values
for i in range(2,20,1):
  CV_Score = cross_val_score(LR,X_rem,y_rem,scoring = 'accuracy',cv = i).mean()
  print('cv = ',i,'Accuracy = ',CV_Score)
```

Maximum value occurs at cv = 6

```python
#Using Cross validation with StratifiedKfold
from sklearn.model_selection import StratifiedKFold
for i in range(2,10,1):
      splits = StratifiedKFold(n_splits = i)
      CV_Score = cross_val_score(LR,X_rem,y_rem,scoring = 'accuracy',cv = splits).mean()
      print('cv = ',i,'Accuracy = ',CV_Score)
```

Here the model gets a maximum accuracy of 0.992704 percent
